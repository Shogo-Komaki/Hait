```python
# 自由度調整済み決定係数
# (決定係数, trainまたはtestのサンプル数, 利用した特徴量の数)
def adjusted(score, n_sample, n_features):
    adjusted_score = 1 - (1 - score) * ((n_sample - 1) / (n_sample - n_features - 1))
    return adjusted_score
print('adjusted R^2')
print('train: %3f' % adjusted(lr.score(X_train, y_train), len(y_train), 2))
print('test : %3f' % adjusted(lr.score(X_test, y_test), len(y_test), 2))
```

```python
Pclass = X.loc[:, [0]].values
Sex = X.loc[:, [1]].values
Age = X.loc[:, [2]].values
y = y.loc[:, ['Survived']].values
```

```python
Pclass
```

```python
#２次関数で回帰
from sklearn.preprocessing import PolynomialFeatures
quad = PolynomialFeatures(degree=2)    # 2次の多項式基底を生成
Pclass_quad = quad.fit_transform(Pclass) # 生成した基底関数で変数変換を実行する
```

```python
from sklearn.model_selection import train_test_split
Pclass_quad_train, Pclass_quad_test, _, _ = train_test_split(Pclass_quad, y, test_size = 0.3, random_state = 0)
```

```python
# 線形回帰のインスタンスを生成
model_quad = LinearRegression()

# 変換後の変数を利用して実行
model_quad.fit(Pclass_quad_train, y_train)
```

```python
# 散布図に近似直線､近似曲線を重ねてみる
x = np.arange(0, 3, 1)[:, np.newaxis] 
plt.scatter(Pclass, y, color='lightgray', label='data')               # 散布図を表示
x_quad = quad.fit_transform(x)                                       # プロットのためのxも2変数変換する
plt.plot(x, model_quad.predict(x_quad), color='green', label='quad') # 回帰曲線を表示（2次関数）
plt.xlabel('Pclass')                                                  # x軸のラベルを表示
plt.ylabel('surived')                                                   # y軸のラベルを表示
plt.legend(loc = 'upper right')                                      # 凡例を表示
plt.show()                                                           # 上記の内容でグラフを表示                               # 上記の内容でグラフを表示
```

```python
#値が離散的なので回帰がうまくいきにくい
```

```python
X_quad = np.hstack((Pclass_quad, Sex, Age))
```

```python
# ホールド・アウト法によるデータの分割
X_quad_train, X_quad_test, _, _ = train_test_split(X_quad, y, test_size = 0.3, random_state = 0)
# 線形回帰による学習
model_quad_2 = LinearRegression()
model_quad_2.fit(X_quad_train, y_train)
```

```python
def adjusted(score, n_sample, n_features):
    adjusted_score = 1 - (1 - score) * ((n_sample - 1) / (n_sample - n_features - 1))
    return adjusted_score
```

```python
# 2次関数
print('model_quad_2')
print('train: %.3f' % adjusted(model_quad_2.score(X_quad_train, y_train), len(y_train), 3))
print('test : %.3f' % adjusted(model_quad_2.score(X_quad_test, y_test), len(y_test), 3))
```

```python
# 3次関数の変数変換
cubic = PolynomialFeatures(degree=3)     # 3次の多項式規定を生成
Pclass_cubic = cubic.fit_transform(Pclass) # 生成した基底関数で変数変換を実行
```

```python
Pclass_cubic_train, Pclass_cubic_test, _, _ = train_test_split(Pclass_cubic, y, test_size = 0.3, random_state = 0)
```

```python
# 線形回帰のインスタンスを生成
model_cubic = LinearRegression()

# 変換後の変数を利用して実行
model_cubic.fit(Pclass_cubic_train, y_train)
```

```python
# 散布図に近似直線､近似曲線を重ねてみる
plt.scatter(Pclass, y, color='lightgray', label='data')                 # 散布図を表示
         
plt.plot(x, model_quad.predict(x_quad), color='green', label='quad')   # 回帰曲線を表示（2次関数）
x_cubic = cubic.fit_transform(x)                                       # プロットのためのxも2変数変換する
plt.plot(x, model_cubic.predict(x_cubic), color='blue', label='cubic') # 回帰曲線を表示（2次関数）
plt.xlabel('Pclass')                                                    # x軸のラベルを表示
plt.ylabel('survived')                                                     # y軸のラベルを表示
plt.legend(loc = 'upper right')                                        # 凡例を表示
plt.show()                                                             # 上記の内容でグラフを表示
```
